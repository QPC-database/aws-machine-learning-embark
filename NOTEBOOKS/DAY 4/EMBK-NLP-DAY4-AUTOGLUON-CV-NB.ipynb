__AutoGluon ImageClassification__ is very easy to use. We just use the following three lines and start fitting our model. We will only train for 3 epochs. Feel free to increase it. It will also apply validation.

```python
from autogluon import ImageClassification as task

dataset = task.Dataset("softImages")

classifier = task.fit(dataset, ngpus_per_trial=1, epochs=3, verbose = 1, 
                      time_limits=600)
```
