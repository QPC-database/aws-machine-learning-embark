#### 2.2 Use AutoGluon ImageClassification to train an image classifier

__AutoGluon ImageClassification__ is very easy to use. We just use the following three lines and start fitting our model. We will only train for 5 epochs. Feel free to increase it. It will also apply validation.

* Find more details on the __AutoGluon ImageClassification__ here: https://autogluon.mxnet.io/api/autogluon.task.html#autogluon.task.ImageClassification


```python
import time
from autogluon import ImageClassification as task

start = time.time()

train_dataset = task.Dataset("images/train")

classifier = task.fit(train_dataset,
                      ngpus_per_trial=1,
                      epochs=5,
                      verbose = 1, 
                      time_limits=600)
end = time.time()

print("Time elapsed: %f Seconds" % (end-start))
```

Let's make a prediction

```python
image = 'images/train/sw/1.jpg'
ind, prob, classes = classifier.predict(image)
print(ind, prob, classes)
```
